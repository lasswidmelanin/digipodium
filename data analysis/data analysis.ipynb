```python
import seaborn as sns

penguins=sns.load_dataset('penguins')
penguins['body_mass_g'].plot(kind='hist',bins=3)
```

```python
def categorize_by_weight(value):
    if value == np.nan:
        return np.nan
    elif value >=2696 and value <= 3900:
        return "light weight"
    elif value > 3900 and value <=5100:
        return "medium weight"
    elif value > 5100:
        return "heavy weight"
# categorize_by_weight(3000)
penguins['category']=penguins['body_mass_g'].apply(categorize_by_weight)
penguins.head()
```
