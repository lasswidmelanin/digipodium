```python
x=[1,2,3,4]
y=[2,3,4,5]
plt.plot(x,y)
plt.show()
```

```python
c=['r','y','g','b']
plt.bar(x,y,color=c)
plt.show()
```

# BAR Plot

```python
x=['python','c','c++','java']
y=[85,70,60,82]
c=['r','y','g','b']
plt.xlabel("languages",fontsize=10)
plt.ylabel("popularity rate",fontsize=10)
plt.title("languages popularity",fontsize=10)
plt.bar(x,y,color=c,width=0.5,align="edge",edgecolor='black',linewidth=2,linestyle=":",alpha=0.5) # by default align="center"
plt.show()
```

```python
import numpy as np


x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title("Languages Popularity Graph")
plt.xlabel("Languages")
plt.ylabel("Popularity Rate")

plt.bar(p,y,label="popularity1",color='r',width=0.2)
plt.bar(p1,z,label="popularity2",color='y',width=0.2)

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```
