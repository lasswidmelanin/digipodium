# BAR Plot

```python
x=['python','c','c++','java']
y=[85,70,60,82]
c=['r','y','g','b']
plt.xlabel("languages",fontsize=10)
plt.ylabel("popularity rate",fontsize=10)
plt.title("languages popularity",fontsize=10)
plt.bar(x,y,color=c,width=0.5,align="edge",edgecolor='black',linewidth=2,linestyle=":",alpha=0.5) # by default align="center"
#alpha lighten the shade of scatter dots and ranges between 0 to 1
plt.show()
```

```python
import numpy as np


x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title("Languages Popularity Graph")
plt.xlabel("Languages")
plt.ylabel("Popularity Rate")

plt.bar(p,y,label="popularity1",color='r',width=0.2)
plt.bar(p1,z,label="popularity2",color='y',width=0.2)

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,30,29,56,20,30] #color code
size=[200,400,100,70,200,500,300]

plt.scatter(day,no,c=colors,s=size,cmap='BrBG',alpha=0.5) # virdis is color category ex- BrBG etc
plt.title('Scatter Plot', fontsize=15)

t=plt.colorbar()
t.set_label('Color Bar',fontsize=15)

plt.xlabel("Day",fontsize=15)
plt.ylabel('No.',fontsize=15)

plt.show()
```

```python
# giving two values in a scatter plot
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,2,4,5,1,6,2]

colors=[10,49,30,29,56,20,30] #color code
size=[200,400,100,70,200,500,300]

plt.scatter(day,no,c=colors,s=size,cmap='BrBG')# virdis is color category ex- BrBG etc
plt.scatter(day,no2,c=colors,s=size,cmap='viridis')
plt.title('Scatter Plot', fontsize=15)

t=plt.colorbar()
t.set_label('Color Bar',fontsize=15)

plt.xlabel("Day",fontsize=15)
plt.ylabel('No.',fontsize=15)

plt.show()
```

# pie chart

```python
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.2,0.0,0.,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=ex,colors=c,
        autopct='%0.1f%%',shadow=True,
        radius=0.8,labeldistance=1.2,
        startangle=0,textprops={"fontsize":13},
        counterclock=False,wedgeprops={'linewidth':4}, #'width':0.2 # 'edgecolor':'m'
        rotatelabels=False #center=(3,3) # rotatelabels=True      
        )

plt.title('Wscube Tech',fontsize=20)
plt.legend(loc=2)
plt.show()
```

# box vs whisker plot

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,
            # vert=False, # vert=False will make horizontal box plot
            # notch=True, # is design of a box  
            # widths=0.1,
             labels=['Python'],
            # patch_artist=True, # colored box
             showmeans=True,
            # whis=3.5,
            sym='g +', # green color and marker shape of outlier
            boxprops=dict(color='r'),
            capprops=dict(color='g'),
            whiskerprops=dict(color='b'),
            # flierprops=dict(markeredge='y')
           )
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]
plt.boxplot(x,
             labels=['Python'],
             showmeans=True,
            sym='g +', # green color and marker shape of outlier
            boxprops=dict(color='r'),
            capprops=dict(color='g'),
            whiskerprops=dict(color='b'),
           )
plt.boxplot(x1,
             labels=['Python'],
             showmeans=True,
            sym='g +', # green color and marker shape of outlier
            boxprops=dict(color='r'),
            capprops=dict(color='g'),
            whiskerprops=dict(color='b'),
           )

plt.show()
```

```python
x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]

y=[x,x1]

plt.boxplot(y,
             labels=['Python','C++'],
             showmeans=True ,
            boxprops=dict(color='r'),
            capprops=dict(color='g'),
            whiskerprops=dict(color='b'),
           )
plt.show()
```
