# Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Line Plot

```python
var=[1,2,3,4,5,6,7]
var1=[2,3,4,1,5,6,7]

df=pd.DataFrame({'var':var , 'var1':var1})

sns.lineplot(x='var',y='var1',data=df)

plt.show()
```

```python
plt.plot(var,var1)
plt.show()
```

```python
var=[1,2,3,4,5,6,7]
var1=[2,3,4,1,5,6,7]

sns.lineplot(x=var,y=var1)
```

```python
d=sns.load_dataset('penguins')
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=d,hue='sex',
             style='sex',   #palette='Accent'
             markers=['o','*'])

plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=d,hue='sex',
             style='sex',   #palette='Accent'
             markers=['o','*'],
             dashes=False,  #giving false joined dash line
             legend=True    #False to remove 

             )

plt.grid()
plt.title("Penguins data")
plt.show()
```
