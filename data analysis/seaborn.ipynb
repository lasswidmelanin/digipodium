# Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Line Plot

```python
var=[1,2,3,4,5,6,7]
var1=[2,3,4,1,5,6,7]

df=pd.DataFrame({'var':var , 'var1':var1})

sns.lineplot(x='var',y='var1',data=df)

plt.show()
```

```python
plt.plot(var,var1)
plt.show()
```

```python
var=[1,2,3,4,5,6,7]
var1=[2,3,4,1,5,6,7]

sns.lineplot(x=var,y=var1)
```

```python
d=sns.load_dataset('penguins')
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=d,hue='sex',
             style='sex',   #palette='Accent'
             markers=['o','*'])

plt.show()
```

```python
d=sns.load_dataset('penguins').head(20)
d
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=d,hue='sex',
             style='sex',   #palette='Accent'
             markers=['o','*'],
             dashes=False,  #giving false joined dash line
             legend=True    #False to remove 

             )

plt.grid()
plt.title("Penguins data")
plt.show()
```

# Histogram plot

```python
peng=sns.load_dataset('penguins')
peng
```

```python
sns.displot(peng['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],
            kde=True,rug=True, #kde= kernel density estimater,rug is to show kde on x axis
            color='g')
plt.show()
```

```python
sns.displot(peng['flipper_length_mm'],
            kde=True)
plt.show()
```

```python
sns.displot(peng['flipper_length_mm'],
            kde=True,rug=True, #kde= kernel density estimater,rug is to show kde on x axis
            color='g',
            log=True) 
plt.show()
```

# Bar plot

```python
peng
```

```python
sns.barplot(x=peng.island,y=peng.bill_length_mm,color='g')
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=peng,hue='sex')
plt.show()
```

```python
order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=peng,hue='sex',
            order=order1, hue_order=['Female','Male'],
            palette='icefire',
            #saturation=0.2 # used as alpha to fade or darken colorbar
            errcolor='y',
            errwidth=5
            # ci=90,n_boot=20
            )
plt.show()
```

```python
order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='bill_depth_mm',y='bill_length_mm',data=peng,
            hue='sex',orient='h')
plt.show()
```

```python
sns.set(style='darkgrid')
order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=peng,hue='sex',
            order=order1, hue_order=['Female','Male'],
            errcolor='b',#alpha=0.5
            #dodge=True/False to remove overlapping
            # or to overlap bars of two categories 
            )
plt.show()
```

# Scatter Plot

```python
peng1=peng.head(20)
peng1
```

```python
mark={'Male':'o','Female':'^'}
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=peng1,hue='sex',
                style='sex',size='sex',sizes=(100,40),palette='gist_ncar',
                #alpha=0.3   
                markers=mark      
                )
plt.show()
```
