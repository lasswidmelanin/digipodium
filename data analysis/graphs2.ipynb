# categorical graphs
- columns that have text are generally categorical
- the first step is to count the number of unique values in column
- then plot the count of each category
- following charts are the good way to visualize categorical data
     - bar chart
     - pie chart
     - donut chart
     - line chart (when too many categories)
     - area chart

```python
print(titanic['survived'].value_counts())
```

```python
survival_count=titanic['alive'].value_counts()
survival_count.plot(kind="pie", autopct="%.1f%%", title='survival rate ')
```

```python
survival_count.plot(kind="pie", 
   autopct="%.1f%%", 
   title='survival rate',
   labels=['Survived','not survived'],
   colors=['#66b3ff','#ff9999'],
   labeldistance=1.1,
   wedgeprops={
       'width':0.5,
   })
```

```python
layer=titanic['pclass'].value_counts()
layer.plot(kind='pie',autopct='%.2f%%',title='titanic layers')
```

```python
layer=titanic['pclass'].value_counts()
layer.plot(kind='bar',title='titanic layers')
```

```python
person=titanic['who'].value_counts()
person.plot(kind='pie',autopct='%.2f%%',title='person gender')
```
